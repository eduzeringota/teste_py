# <font color='black'>Simulação de Dataset</font>

### <font color='red'>Parâmetros</font>

```python
rd.seed = 2020

qtd_entidades = 10

qtd_cat_servicos_min = 2
qtd_cat_servicos_max = 4

qtd_servicos_min = 6
qtd_servicos_max = 20

qtd_acoes_min = 4
qtd_acoes_max = 9
```

### <font color='red'>Entidades</font>

```python
dic_entidades = {}
for i in range(qtd_entidades):
    x = i + 1
    dic_entidades[x] = "Entidade " + str(x)
```

```python
dic_entidades
```

### <font color='red'>Serviços</font>

```python
dic_servicos = {}
idx = 0
for i in dic_entidades:
    n_serv = rd.randrange(qtd_servicos_min, qtd_servicos_max)
    for k in range(n_serv):
        n_cat_serv = rd.randrange(qtd_cat_servicos_min, qtd_cat_servicos_max)
        x = k + 1
        idx += 1
        dic_servicos[idx] = [dic_entidades[i],
                             "Servico " + str(x) + " da " + dic_entidades[i],
                             "Categoria de Servico " + str(n_cat_serv)]
```

```python
dic_servicos
```

### <font color='red'>Ações</font>

```python
dic_acoes = {}
idx = 0
for i in dic_servicos:
    n_acoes = rd.randrange(qtd_acoes_min, qtd_acoes_max)
    for k in range(n_acoes):
        x = k + 1
        idx += 1
        dic_acoes[idx] = [dic_servicos[i][0],
                          dic_servicos[i][1],
                          dic_servicos[i][2],
                          "Acao " + str(x) + " do " + dic_servicos[i][1]]
```
